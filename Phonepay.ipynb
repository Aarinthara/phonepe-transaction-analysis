```python
# Data Consistency Check Between State and District Levels
```

```python
# District Level Aggregation

district_summary = district_Txn_and_Users.groupby(["State", "Year", "Quarter"])[["Transactions", "Amount (INR)", "Registered Users"]].sum()
district_summary.reset_index()
```

```python
# State Level Summary

state_summary = state_Txn_and_Users[["State", "Year", "Quarter","Transactions", "Amount (INR)", "Registered Users"]]
state_summary
```

```python
# Merge State and District Summaries

comparison_data = pd.merge(district_summary.reset_index(),state_summary,on=["State", "Year", "Quarter"],suffixes=("_District", "_State"))
comparison_data
```

```python
# Differences

comparison_data["Transaction_Difference"] = (comparison_data["Transactions_State"]-comparison_data["Transactions_District"])

comparison_data["Amount_Difference"] = (comparison_data["Amount (INR)_State"]- comparison_data["Amount (INR)_District"])

comparison_data["Users_Difference"] = (comparison_data["Registered Users_State"]- comparison_data["Registered Users_District"])

comparison_data
```

```python
# Display Discrepancies

discrepancies = comparison_data[(comparison_data["Transaction_Difference"] != 0) |(comparison_data["Amount_Difference"] != 0)|(comparison_data["Users_Difference"] != 0)]

discrepancies
```

-- Observation
District-level transaction totals and registered user counts
closely match the corresponding state-level data.

Small differences were observed in transaction amounts,
mainly due to decimal precision and rounding variations
during aggregation calculations.

Overall, the dataset shows good consistency and reliability
between district-level and state-level records.

--Observation
The user-to-population ratio differs significantly across states,
showing variations in digital payment adoption and PhonePe usage.

Delhi, Telangana, and Dadra & Nagar Haveli and Daman & Diu
recorded comparatively higher ratios, indicating stronger
app penetration and user engagement.

States with lower ratios may have comparatively lower
digital adoption or smaller active user bases.

# Summary of Findings and Insights

This project analyzed PhonePe transaction and demographic data across various states and districts in India using Python, Pandas, Matplotlib, and Seaborn.

The analysis identified strong growth in digital transactions and app usage from 2018 to 2021, indicating increasing adoption of digital payment platforms across the country.

States such as Karnataka, Maharashtra, Telangana, Andhra Pradesh, and Tamil Nadu recorded higher transaction activity and stronger user engagement. Peer-to-peer payments and merchant payments contributed the largest share of overall transactions.

The project also revealed that urban and densely populated regions generally showed higher transaction volumes and stronger digital payment adoption. Smartphone brands such as Xiaomi, Samsung, Vivo, and Oppo dominated the user base across multiple states.

Correlation analysis indicated a moderate positive relationship between population density and transaction activity, suggesting that urbanization plays an important role in digital payment growth.
